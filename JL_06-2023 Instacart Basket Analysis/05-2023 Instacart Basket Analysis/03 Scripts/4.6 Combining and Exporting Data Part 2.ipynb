## Step 5: Determine a way to combine the orders_products_combined dataframe with the products dataset.

```python
#Check columns for df_prods

df_prods.head()
```

```python
#Check columns for dr_ords_prods_combined

df_ords_prods_combined.head()
```

```python
# Drop unneeded column in df_ords_prods_combined: '_merge'

df_ords_prods_combined = df_ords_prods_combined.drop(columns = ['_merge'])
```

```python
#Check to see if it was dropped properly 

df_ords_prods_combined.head()
```

```python
# Use column 'product_id' as key to merge the dataframes

df_merged = df_ords_prods_combined.merge(df_prods, on = 'product_id', indicator = True)
```

```python
#Check df_merged

df_merged.head()
```

```python
#Drop 3 unnecessary columns: 'Unnamed: 0.1', 'Unnamed: 0_x', and 'Unnamed: 0_y'

df_merged = df_merged.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0_x', 'Unnamed: 0_y'])
```
