```python
#Define the function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

## If statements with the loc() Function

```python
#First condition: High range if price over 15

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#Second condition: Mid range if price is between 5 and 15

df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
#Third condition: Low range if price is under 5

df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Count values to see if it matches user-defined function without loc()

df['price_range_loc'].value_counts(dropna = False)
```

### Make for loop to summarize how busy each day of the week is

```python
# Count frequency of orders on each day of the week

df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
#Function with the for-loop

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

# Step 2: Create a new column that has the following criteria:
## "Busiest Days" matches the two busiest days of the week
## "Slowest Days" matches the two slowest days of the week
## "Regularly Busy" matches the others

```python
#Check to see which days match the busiest days and slowest days

df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
#Create new function with the for-loop

result_2 = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest Days")
  else:
    result_2.append("Regularly busy")
```

```python
#Check the new result

result_2
```

```python
#Set a new column for the results in the new for-loop

df_ords_prods_merged['busiest_days'] = result_2
```

# Step 4: Find the busiest hours of the day in groups with "Most orders", "Average orders" and "Fewest orders". Create a new column with these labels called "busiest_period_of_day".

```python
#Look to see which column will be needed.

df_ords_prods_merged.head()
```

```python
# Use order_hour_of_day values. Check frequency of the order hours. 

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

### The 24 listed hours can be split up into 3 groups, with the top 8 being "Most orders", middle 8 being "Average orders", and last 8 being "Fewest orders"

```python
# New Function with the for-loop

result_3 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
    result_3.append("Average orders")
  else:
    result_3.append("Fewest orders")
```

```python
# Check the new result

result_3
```

```python
#Set a new column for the results in the new for-loop

df_ords_prods_merged['busiest_period_of_day'] = result_3
```
