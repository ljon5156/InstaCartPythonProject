# Step 6: Create a spending flag for each user based on their average price across all their orders. If mean of their products is less than 10, flag as a "Low Spender". Flag them as a "High Spender" if their price is 10 or higher.

```python
#Group each user based on their user_id, then use transform() to find the average of 'prices' and create a new column called
#'average_price'.

ords_prods_merged['average_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Check to see the new grouping

ords_prods_merged.head(20)
```

```python
# Use 'average_price' column to create the flag for Low or High Spender

ords_prods_merged.loc[ords_prods_merged['average_price'] >= 10, 'spending_status'] = 'High Spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['average_price'] < 10, 'spending_status'] = 'Low Spender'
```

```python
#Check frequency of values

ords_prods_merged['spending_status'].value_counts(dropna = False)
```

# Step 7: Create an order frequency flag that marks a user's ordering behavior according to the median in the "days_since_prior_order" column. "Non-frequent customers" have a median higher than 20, "Regular_customers" have a media greater than 10 but up to 20, "Frequent Customers" have a median of 10 or lower.

```python
#Group customers by 'user_id' first, then use transform to find the median of 'days_since_last_order' and create a new 
#column called median_days_since_last_order.

ords_prods_merged['median_days_since_last_order'] = ords_prods_merged.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
#check to see if new column was made correctly

ords_prods_merged.head(20)
```

```python
#Create a flag for the 'median_days_since_last_order' column with the criteria

ords_prods_merged.loc[ords_prods_merged['median_days_since_last_order'] > 20, 'customer_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_days_since_last_order'] > 10) & (ords_prods_merged['median_days_since_last_order'] <= 20), 'customer_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['median_days_since_last_order'] <= 10, 'customer_frequency_flag'] = 'Frequent customer'
```

```python
#Check frequency for the flag

ords_prods_merged['customer_frequency_flag'].value_counts(dropna = False)
```
