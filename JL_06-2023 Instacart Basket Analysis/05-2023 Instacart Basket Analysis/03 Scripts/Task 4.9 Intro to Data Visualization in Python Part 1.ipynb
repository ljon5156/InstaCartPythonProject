# Step 5: Data quality and consistency checks on the Customers dataset

## Mixed Data Type Columns

```python
#Check for mixed-types

for col in df_customers.columns.tolist():
    weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_customers[weird]) > 0:
        print(col)
```

#### The only mixed-data column is the 'first_name' column which is because of the 'NaN's' in this column which were found in Step 4. The rest of the column are 'strings', which is appropriate for a name column so the column's datatype should be turned into a 'string'.

```python
#Change the mixed data column into a singular type of data

df_customers['first_name'] = df_customers['first_name'].astype('str')
```

```python
#Check again to see if there are any mixed-types

for col in df_customers.columns.tolist():
    weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_customers[weird]) > 0:
        print(col)
```
